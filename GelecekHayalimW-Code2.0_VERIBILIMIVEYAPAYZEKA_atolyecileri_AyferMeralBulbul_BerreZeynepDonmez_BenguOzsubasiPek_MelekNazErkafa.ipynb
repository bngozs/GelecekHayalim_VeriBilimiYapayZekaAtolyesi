**2. Aşama: Veri Setinin İstatistiksel Analizi**

***İstatistiksel Analiz***

TYT Puan Dağılımı: Medyan ve ortalama birbirine yakın olup normal dağılımı göstermektedir. Ancak mod, ortalamanın önemli ölçüde altındadır. Düşük puanların kümelenmesi söz konusudur.

AYT Puan Dağılımı: TYT'nin aksine, mod, medyan ve ortalama birbirine yakın olup yine normal dağılımı göstermektedir. Ancak başarı oranı daha yüksek olup öğrenciler konu bazlı derslerde daha iyi performans göstermektedir.

Lise Mezuniyet Ortalaması: TYT'ye benzer. Öğrencilerin TYT'deki başarısı, lise sınavlarındaki başarılarıyla ilişkilidir.

Çalışma Saatleri: Normal dağılım göstermekte ve genellikle günde ortalama 6 saat olarak belirlenmiştir.

Temel istatistiksel analize göre, TYT eksikliklerini ve düşük puan birikimini gidermek için çalışma saatlerini niceliksel olarak artırmak yerine niteliksel bir stratejiye ihtiyaç duyulduğu tahmin edilmektedir.

Gerekli dizilerin benzersiz elemanları NumPy kullanılarak bulunmuştur.

Gruplandırma işlemi gerçekleştirildi ve kategoriler arasındaki ilişkiler analiz edildi.

```python
kritik_veri = ['TYT Puanı', 'AYT Puanı', 'Lise Mezuniyet Ortalaması', 'Günlük Çalışma Süresi','Haftalık Çalışma Süresi']

ist_ozet = {}

for col in kritik_veri:
   ist_ozet[col] = {
        'Ortalama': df[col].mean(),
        'Medyan': df[col].median(),
        'Mod':df[col].mode()[0],
        'Std': df[col].std(),
        'Min': df[col].min(),
        'Max': df[col].max()
    }

df_ist = pd.DataFrame(ist_ozet).T.round(2)

print("--- YKS 2025 TEMEL İSTATİSTİKLER---")
print(df_ist)
```

```python
kiz_ogrenciler = df[df['Cinsiyet'] == 'Kadın'].copy()

bolge_toplam_kiz = kiz_ogrenciler['Bölge'].value_counts()

stem_no_doctor_condition = (kiz_ogrenciler['Alan'] == 'Sayısal') & (kiz_ogrenciler['Meslek Hedefi'] != 'Doktor')
bolge_stem_kiz = kiz_ogrenciler[stem_no_doctor_condition]['Bölge'].value_counts()

stem_orani_kiz = (bolge_stem_kiz.reindex(bolge_toplam_kiz.index, fill_value=0) / bolge_toplam_kiz) * 100
stem_orani_kiz = stem_orani_kiz.sort_values(ascending=False)

tum_ogrenci_sayisi = len(df)

genel_stem_no_doctor = df[(df['Alan'] == 'Sayısal') & (df['Meslek Hedefi'] != 'Doktor')]
tum_stem_ogrenci_sayisi = len(genel_stem_no_doctor)

turkiye_genel_ort = (tum_stem_ogrenci_sayisi / tum_ogrenci_sayisi) * 100

plt.figure(figsize=(12, 7))
plot_data = stem_orani_kiz.reset_index()
plot_data.columns = ['Bölge', 'STEM Oranı']

ax = sns.barplot(x='Bölge', y='STEM Oranı', data=plot_data, palette='Blues_d')

for p in ax.patches:
    ax.text(x=p.get_x() + p.get_width()/2,
            y=p.get_height() + 0.5,
            s=f"%{p.get_height():.1f}",
            ha='center', fontsize=12, fontweight='bold')


plt.axhline(y=turkiye_genel_ort, color='red', linestyle='-', linewidth=2,
            label=f"TR Geneli : %{turkiye_genel_ort:.1f}")

plt.title('Bölgelere Göre Kız Öğrencilerin Mühendislik/Teknoloji (STEM) Tercihi)', fontsize=14, fontweight='bold')
plt.xlabel('Bölge', fontsize=12)
plt.ylabel('Tercih Oranı (%)', fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, max(stem_orani_kiz.max(), turkiye_genel_ort) + 10)
plt.legend(loc='upper right', frameon=True, facecolor='white', framealpha=0.9)
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```
