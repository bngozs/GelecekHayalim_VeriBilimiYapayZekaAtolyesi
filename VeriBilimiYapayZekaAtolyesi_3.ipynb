```python
# Odev 3- Alistirma 1 Daire Alanı Hesaplayıcı
# Kullanıcıdan pi değeri ve yarıçap bilgisi alarak dairenin alanını hesaplayan bir fonksiyon oluşturulur.

def daire_alan_hesabi(pi,yaricap):
    alan = pi*(yaricap**2)
    return alan

pi = float(input("Lütfen pi sayı değerini giriniz="))
yaricap = float(input("Lütfen dairenin yarıçap değerini giriniz="))

daire_alani = daire_alan_hesabi(pi,yaricap)

print(f"Yarıçapı {yaricap} olan dairenin alanı=",{daire_alani})
```

```python
# Odev 3- Alistirma 2 Faktöriyel Hesaplayıcı
# Faktöriyel adında fonksiyon oluşturulur. Döngü kullanarak parametre olarak girilen sayının faktöriyeli hesaplanır. 
# Format metodunu kullanılarak ekrana yazdırılır.

def faktoriyel(sayi):
    sonuc = 1
    for i in range (1, sayi + 1):
        sonuc = sonuc * i
    return sonuc

sayi1 = int(input("Faktöriyeli alınmasını istediğiniz sayıyı giriniz="))
hesap = faktoriyel(sayi1)

print(f"{sayi1} faktöriyel değerinin sonucu=",hesap)
```

```python
# Odev 3- Alistirma 3 Yaş Hesaplayıcı
# Kişinin fonksiyona doğum yılını vererek kaç yaşında olduğunu hesaplayan bir fonksiyon oluşturun. 

def yas_hesabi(dogum_yili,yil):
    yas = yil - dogum_yili
    return yas

dogum_yili = int(input("Doğum yılınızı giriniz="))
yil = int(2025)
hesap = yas_hesabi(dogum_yili,yil)

print(f"Yaşınız=", hesap)
```

```python
# Odev 3- Alistirma 4 Emeklilik Hesaplayıcı
# Doğum yılı ve isim bilgisi verilen fonksiyon kişinin emekli olup olmadığını söylesin.  
# Kişi 65 yaşında ya da daha fazlaysa "Emekli oldunuz" yanıtını, 
# 65 yaşından küçükse emekliliğine kaç yıl kaldığını da hesaplayarak "(isim) emekliliğine (yıl) kaldı." yanıtını versin.

def yas_hesabi(dogum_yili,yil):
    yas = yil - dogum_yili
    return yas

isim = input("İsminizi giriniz=")
dogum_yili = int(input("Doğum yılınızı giriniz="))
yil = int(2025)
hesap = yas_hesabi(dogum_yili,yil)

if hesap >= 65:
    print(f"Emekli oldunuz!")
else:
    hesap2 = 65 - hesap
    print(f"Emekliliğinize {hesap2} yıl kalmıştır.")
```
