```python
# Odev 1 - Alistirma 2:  İsimlerden oluşan üç değişkene yaş değerleri atanır. 
# Belirlenen üç değişken birbiriyle karşılaştırma operatörleri ile karşılaştırılır. 
# Bu karşılaştırmalara mantıksal operatörler de eklenir.
```

```python
Eren_Yasi = 25
Seren_Yasi = 22
Selim_Yasi = 26

Sonuc = Eren_Yasi < Seren_Yasi and Selim_Yasi > Eren_Yasi

print("Eren, Seren ve Selim'den küçük müdür?", Sonuc)
```

```python
# Odev 1- Alistirma 3:  Kullanıcıdan iki değer girmesini istenir. 
# Girilen değerlerin toplama, çıkarma, çarpma, bölme sonuçlarını yazdırılır.

sayi1 = input("1.sayıyı giriniz:")
sayi2 = input("2.sayıyı giriniz:")

sayi1_float = float(sayi1)
sayi2_float = float(sayi2)


Toplama = (sayi1_float + sayi2_float)
Cikarma = (sayi1_float - sayi2_float)
Carpma = (sayi1_float * sayi2_float)
Bolme = (sayi1_float / sayi2_float)

print("1.sayı + 2.sayı=",Toplama)
print("1.sayı - 2.sayı=",Cikarma)
print("1.sayı x 2.sayı=",Carpma)
print("1.sayı / 2.sayı=",Bolme)
```

```python
# Odev 1- Alistirma-4: Kullanıcıdan isim, yaş, şehir ve meslek bilgilerini istenir. 
# Cevaplarını yazdırılır.

isim = input("İsminizi Giriniz:")
yas = input("Yaşınızı Giriniz:")
sehir = input("Yaşadığınız Şehri Giriniz:")
meslek = input("Mesleğinizi Giriniz:")

print("İsim:",isim)
print("Yaş:",yas)
print("Şehir:",sehir)
print("Meslek:",meslek)
```

```python
# Odev 1- Alistirma-5:  "Hi-Kod Veri Bilimi Atölyesi" ifadesini bir değişkene tanımlanır.
# İfadedeki her bir kelimeyi ("Hi-Kod", "Veri", "Bilimi", "Atölyesi") değişken içinden seçilir.
# İfadeyi hepsini büyük harf olacak hale çevrilir. ("HI-KOD VERİ BİLİMİ ATÖLYESİ")
#İfadeyi hepsini küçük harf olacak hale çevrilir.("hi-kod veri bilimi atölyesi")
# "0123456789" ifadesindeki yalnızca çift sayıları ve yalnızca tek sayıları seçilir. 
#("02468", "13579")

Cumle = "Hi-Kod Veri Bilimi Atölyesi"
Kelimeler = Cumle.split()

print("Cümlenin Tamamı:", Cumle)
print("Cümlenin Kelimeleri:", Kelimeler)

Cumle_BuyukHarf = Cumle.upper()
print("Büyük harf ile cümlenin yazılışı:", Cumle_BuyukHarf)

Cumle_KucukHarf = Cumle.lower()
print("Küçük harf ile cümlenin yazılışı:", Cumle_KucukHarf)

Sayilar = "0123456789"
Cift_Sayilar = Sayilar [0::2]
Tek_Sayilar = Sayilar [1::2]
print("Çift sayılar=",Cift_Sayilar)
print("Tek sayılar=",Tek_Sayilar)
```
